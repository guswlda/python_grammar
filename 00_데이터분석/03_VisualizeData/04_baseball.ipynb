```python
import pandas as pd

# 데이터 로드
atbat = pd.read_csv("../00_data/atbats.csv")
pitch = pd.read_csv("../00_data/pitches.csv")
name = pd.read_csv("../00_data/player_names.csv")

# atbat과 pitch를 ab_id 기준으로 병합
merged_data = pd.merge(atbat, pitch, on="ab_id", how="inner")

# 병합된 데이터와 name을 first_name 기준으로 다시 병합
totalData = pd.merge(merged_data, name, left_on="first_name", right_on="id", how="inner")

# 결과 확인
print(totalData)
```

```python
import pandas as pd
import math

# 거리를 계산하는 함수
def calculate_distance(row):
    return math.sqrt(row['px']**2 + row['pz']**2)

filtered_data = totalData[totalData['type'].isin(['S'])]

# 'distance' 열에 거리 값 저장
filtered_data['distance'] = filtered_data.apply(calculate_distance, axis=1)

# 결과 출력, 'pitcher_id'와 'distance' 열만 선택하여 반올림
print(filtered_data[['pitcher_id', 'distance']].round(2))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import math

# 거리를 계산하는 함수
def calculate_distance(row):
    return math.sqrt(row['px']**2 + row['pz']**2)

filtered_data = totalData[totalData['type'].isin(['S'])]

# 'distance' 열에 거리 값 저장
filtered_data['distance'] = filtered_data.apply(calculate_distance, axis=1)

# 결과 출력, 'pitcher_id'와 'distance' 열만 선택하여 반올림
# print(filtered_data[['pitcher_id', 'distance']].round(2))

def plot_strike_zone(first_name):
    # 특정 투수 이름에 해당하는 데이터만 필터링
    pitcher_data = filtered_data[filtered_data['first_name'] == first_name]

    # 투수의 투구량 계산
    pitch_count = len(pitcher_data)

    # 그래프 그리기
    plt.figure(figsize=(6, 6))
    scatter = plt.scatter(pitcher_data['px'], pitcher_data['pz'], c=pitcher_data['distance'])
    plt.colorbar(scatter, label='Distance from Center')
    plt.xlim(-2, 2)
    plt.ylim(0, 5)
    plt.title(f"{first_name} - Total Pitches: {pitch_count}")
    plt.xlabel("Strike Zone")
    plt.ylabel("Distance")
    plt.grid(True)
    plt.show()
```

```python
# 특정 투수의 피치 패턴과 위치, 그리고 스트라이크 존 내에서의 투구 거리 분포를 시각적으로 분석
plot_strike_zone('John')
```
