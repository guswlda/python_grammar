# DataFrame
pandas 라이브러리에서 제공하는 가장 중요한 데이터 구조 중 하나로, 2차원 데이터 구조이다. <br>
즉, 행(row)와 열(colum)으로 구성된 테이블 형태의 데이터를 다룰 수 있다. <br>
엑셀 시트나 SQL 테이블과 유사한 형태로 데이터를 처리할 수 있다. <br>
- 데이터 프레임의 주요 특징 :
  - 행(row)와 열 (column)으로 이루어진 2차원 데이터 구조이다.
  - 각 열을 서로 다른 데이터 타입을 가질 수 있다.
  예를 들어, 숫자, 문자열, 날짜 등의 데이터를 하나의 데이터프레임 안에서 처리할 수 있다.
  - 인덱스(index) : 각 행에는 고유한 인덱스가 있으며, 인덱스를 통해 데이터를 빠르게 검색하거나 접근할 수 있다.
  - 데이터 분석, 변형, 정리 등 다양한 작업을 쉽게 할 수 있다.

### 필터링 옵션

```python
stock_df = pd.read_csv("../00_data/stock_daily_prices.csv")
stock_df
```

```python
# stock_df 에서 appl의 주가가 440달러를 넘은 시점을 모두 조회하기
aapl = stock_df[(stock_df["AAPL"]>=440.25000)]
aapl
```

```python
stock_df[(stock_df["Date"] == "2020-08-05")]
```

```python
# .isin
# 특정 열에 해당 값이 존재하는 지 확인하여 반환하는 메서드
mask = stock_df["Date"].isin(["2012-01-19", "2020-08-06"])
stock_df[mask]
```

```python
# .between(), ( = SQL)
# 범위 연산을 하는 메서드이다. (날짜 연산 시 주로 사용)
stock_df[stock_df["Date"].between("2012-01-19", "2013-08-06")]
```

```python
# .duplicated
# 중복되는 행을 찾는 메서드
# keep은 모든 중복되는 값을 true로 표시하고 그렇지 않은 값은 False이다

# print(stock_df["AAPL"].round())
# 중복 될 경우 반환
mask_true = stock_df["AAPL"].round().duplicated(keep=False)
print(stock_df[mask_true])

# 중복 X 반환
mask = ~stock_df["AAPL"].round().duplicated(keep=False)
print(stock_df[mask])
```

```python
order_df = pd.read_csv("../00_data/order_data.csv")
order_df
```

```python
# .drop_duplicates
# 데이터 프레임에서 중복된 행을 제거 시 사용하는 메서드
order_df.drop_duplicates(subset=["Customer_Name"],inplace=True)
order_df
```

```python
# where
# 데이터 프레임에서 필터링을 하기 위한 메서드
# 기준을 축종하지 않는 모든 행은 NaN으로 설정
# 이것은 전체 데이터를 기준으로 결측값을 확인하기 위해서 주로 사용
mask = order_df['Price_per_Unit'] >= 300
order_df.where(mask)
```

```python
# isnull()
# 데이터 프레임이나 시리즈에서 결측값을 확인하는데 사용하는 메서드이다.
order_df.where(mask).isnull()
```

```python
order_df.where(mask).isnull().sum()
```

```python
# dropna
# 데이터 프레임의 결측값(Nan)을 포함하는 행이나 열을 제거하는 데 사용되는 메서드
# how = "any" 행 또는 열 중 하나라도 겨측값이 존재하는 경우 해당 행을 제거
# how = "all" 모든 값이 결측인 경우에만 해당 행을 제거
print(order_df.isnull().sum())
result = order_df.dropna(how="any")
print(result.isnull().sum())

result2 = order_df.where(mask).dropna(how="all")
print(result2.isnull().sum())
```

```python
# .mean()
# 평균을 구하는 메서드
order_df["Quantity"].mean()
```

```python
# fillna()
# 데이터 프레임 혹은 시리즈에서 결측값을 지정한 값으로 채우는데 사용하는 메서드

result = order_df.fillna({"Product" : "fillnaProduct"})
result
```
